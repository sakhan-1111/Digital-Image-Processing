## 3. From the “04-Image-Enhacement” slides, translate and apply the Gray Level Slice method to lena-gray. The resultant image should look like below: 


```python
# Importing libraries
from PIL import Image
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Defining function from gray level sclicing
def glslice(f,A,B,T1,T2):
    image=cv2.imread(f,0)
    x,y=image.shape
    z=np.zeros((x,y))
    for i in range(0,x):
        for j in range(0,y):
            if(image[i][j]>A and image[i][j]<B):
                z[i][j]=T1
            else:
                z[i][j]=T2
    concat = np.hstack((image,z))
    return concat
```

```python
# Calling the defined function
img = glslice('lena_gray.jpg', 125, 205, 255, 0)

# Showing the sliced image
plt.title('Original and Gray level sliced image')
plt.imshow(img,'gray')
plt.show()
```
