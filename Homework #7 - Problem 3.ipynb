3. Write Python code that performs a JPEG (lossy) compression and does the following:
    - Read the Lena-colored image.
    - Apply JPEG compression to the image at 15%.
    - Restore the image from the compression.
    - Plot the original and compressed images side-by-side.
    - Plot the histograms of the original and compressed images side-by-side.
    - Print the file sizes of the original, compressed, and restored images.

```python
# Importing libreries
import numpy as np
import cv2
import matplotlib.pyplot as plt
import os

# Reading lena image
lena = cv2.imread('lena_color.png')

# Compression parameters
params = [cv2.IMWRITE_JPEG_QUALITY, 15]

# JPEG compression
retval, buffer = cv2.imencode('.jpg', lena, params)

# Decode into a numpy array
lena_compressed = buffer

# Restore the image
lena_res = cv2.imdecode(buffer, cv2.IMREAD_COLOR)

# Saving the compressed & restored images.
cv2.imwrite('lena_compressed.jpg', lena_compressed)
cv2.imwrite('lena_restore.jpg', lena_res)

# Display the original and compressed images side-by-side.
cv2.imshow('Original images', lena)
cv2.imshow('Compressed images', lena_compressed)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)

# Plot of the histograms of the original and compressed images side-by-side.
fig, (ax1, ax2) = plt.subplots(1, 2)
fig.suptitle('Histograms of the original and compressed image side-by-side')
ax1.hist(lena.ravel(), bins=256, range=(0, 256))
ax2.hist(lena_compressed.ravel(), bins=256, range=(0, 256))
plt.show()

# Print the file sizes.
original = os.path.getsize('lena_color.png')
compressed = os.path.getsize('lena_compressed.jpg')
restore = os.path.getsize('lena_restore.jpg')

print(f'Original size: {original} bytes')
print(f'Compressed size: {compressed} bytes')
print(f'Restored size: {restore} bytes')
```
