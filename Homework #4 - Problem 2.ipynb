## 2. Apply the Mixed Filtered process using the Prewitt operator and generate the eight images of the lena-gray image from HW3. 

```python
# Importing libraries
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
from PIL import Image
import PIL.ImageOps    
import argparse
import math
import scipy.ndimage
import imageio as iio
import matplotlib.image as mpimg
```

```python
# Importing "cell020004.jpg" image
s1 = Image.open('lena_gray.jpg')

# Printing imported image
plt.imshow(s1)
plt.show()
```

```python
# Sharpening original image
img = cv2.imread('lena_gray.jpg', 1)

# Preparing kernel for sharpening filter
kernel = np.array([[-1, -1, -1],
                 [-1, 8, -1],
                 [-1, -1, -1]])

# Applying sharpening filter
s2 = cv2.filter2D(img, -1, kernel)

# Showing sharpened image
plt.imshow(s2)
plt.show()
```

```python
# adding images
s3 = s1 + s2

cv2.imwrite(r's3.png', s3)
s3 = cv2.imread('s3.png', 1)

# Shwing added image
plt.imshow(s3)
plt.show()
```

```python
# prewitt filtering
img = cv2.imread('lena_gray.jpg')
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gaussian = cv2.GaussianBlur(gray,(3,3),0)

# prewitt
kernelx = np.array([[1,1,1],[0,0,0],[-1,-1,-1]])
kernely = np.array([[-1,0,1],[-1,0,1],[-1,0,1]])

img_prewittx = cv2.filter2D(img_gaussian, -1, kernelx)
img_prewitty = cv2.filter2D(img_gaussian, -1, kernely)

s4 = img_prewittx + img_prewittx

# Showing prewitt filtered image
plt.imshow(s4, 'gray')
plt.show()
```

```python
# Adding ones to image
ones = np.ones((5,5))/25

# Lowpass filter
s5 = scipy.ndimage.correlate(s4, ones, mode='constant')

cv2.imwrite(r's5.png', s5)
s5 = cv2.imread('s5.png', 1)

# Showing added image
plt.imshow(s5)
plt.show()
```

```python
# Multipling images
mul_1 = np.double(s3)/255
mul_2 = np.double(s5)/255
s6 = np.array((mul_1*mul_2)*255, dtype = 'uint8')

# Showing resulting image
plt.imshow(s6)
plt.show()
```

```python
# Adding S1 and S6
s7 = s1 + s6

# Showing added image
plt.imshow(s7)
plt.show()
```

```python
# Normalizing image
img_normalized = cv2.normalize(s7, None, 0, 0.5, cv2.NORM_MINMAX, dtype=cv2.CV_32F)

# Showing normalized image
plt.imshow(img_normalized)
plt.show()

img_normalized = np.asarray(img_normalized)
```

```python
# Gamma correction

gamma = 0.25

s8 = np.array(255*(img_normalized / 255) ** gamma, dtype = 'uint8')

# Showing gamma corrected image 
plt.imshow(s8)
plt.show()
```

```python
# Showing all the images together
plt.subplot(4, 4, 1)
plt.imshow(s1)
plt.subplot(4, 4, 2)
plt.imshow(s2)
plt.subplot(4, 4, 3)
plt.imshow(s3)
plt.subplot(4, 4, 4)
plt.imshow(s4, 'gray')
plt.subplot(4, 4, 5)
plt.imshow(s5)
plt.subplot(4, 4, 6)
plt.imshow(s6)
plt.subplot(4, 4, 7)
plt.imshow(s7)
plt.subplot(4, 4, 8)
plt.imshow(s8)
plt.show()
```
