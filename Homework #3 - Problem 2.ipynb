## 2. Restore the image “lena_noisy.png.” Download the image from the module.

```python
# Importing libraries
import cv2
import numpy as np
from skimage import io
from skimage.restoration import denoise_bilateral
```

```python
# Importing "lena_noisy.png" image
lena_noisy = io.imread("lena_noisy.png")

# Printing imported image
io.imshow(lena_noisy)
```

```python
# Applying denoise bilateral from skimage.restoration 5 times
denoised_image_1 = denoise_bilateral(lena_noisy, channel_axis=2)
denoised_image_2 = denoise_bilateral(denoised_image_1, channel_axis=2)
denoised_image_3 = denoise_bilateral(denoised_image_2, channel_axis=2)
denoised_image_4 = denoise_bilateral(denoised_image_3, channel_axis=2)
denoised_image_5 = denoise_bilateral(denoised_image_4, channel_axis=2)

# Showing denoised image
io.imshow(denoised_image_5)
```

```python
# Making kernel for sharpening
kernel = np.array([[0, -1, 0],
                   [-1, 5,-1],
                   [0, -1, 0]])

# Using opencv for shapening the denoised image
image_sharp = cv2.filter2D(src=denoised_image_5, ddepth=-1, kernel=kernel)

# Showing the final image.
io.imshow(image_sharp)
```
