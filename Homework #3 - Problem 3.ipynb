## 3. Use the image “lena_gray.jpg.” Download the image from the module. Apply the blur, median, and a highboost with A=1.2. Display the 2x2 image with all four images.

```python
# Importing libraries
import cv2
import numpy as np
from skimage import io
import matplotlib.pyplot as plt
```

```python
# Importing "lena_gray.jpg" image
lena = io.imread("lena_gray.jpg")

# Printing imported image
io.imshow(lena)
```

```python
# Blur with 9x9 mask
lena_blur = cv2.blur(lena,(9,9))

# Showing blurred image
io.imshow(lena_blur)
```

```python
# Median filter with 9x9 mask
lena_median = cv2.medianBlur(lena, 9)

# Showing median filtered image
io.imshow(lena_median)
```

```python
# Defining function to perform highboost filter on an image
def highBoostFiltering(image,boost_factor):
    """Function to perform highboost with given boost_factor"""
#     image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) #Converting Image to Gray Scale
    resultant_image = image.copy()
    for i in range(1,image.shape[0]-1):
        for j in range(1,image.shape[1]-1):
            blur_factor = (image[i-1, j-1] + image[i-1, j] - image[i-1, j+1] + image[i, j-1] + image[i, j] + image[i, j+1] + image[i+1, j+1] + image[i+1, j] + image[i+1, j+1])/9
            mask = boost_factor*image[i, j] - blur_factor
            resultant_image[i, j] = image[i, j] + mask
            
    return resultant_image
```

```python
# Highboost filter with A = 1.2
lena_highboost = highBoostFiltering(lena, 1.2)

# Showing highboost filtered image
io.imshow(lena_highboost)
```

```python
# Using subplot to show all 4 images together
f, ax = plt.subplots(2,2)
ax[0,0].imshow(lena)
ax[0,1].imshow(lena_blur)
ax[1,0].imshow(lena_median)
ax[1,1].imshow(lena_highboost)
```
